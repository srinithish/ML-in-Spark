```python
from pyspark import SparkContext
from pyspark.ml.linalg import DenseVector
from pyspark.ml.feature import StandardScaler


filename="Google.csv"
spark = SparkSession.builder.getOrCreate()
google = spark.read.load(filename , format = 'csv',header = 'True', inferschema = 'true', sep= ",")
print("# of rows in data:",google.count())
google.take(3)


```

<h3> Part 2: </h3> Tokenizing and Removing Stop Words

```python
import re

#Reading stop word file
stopwords = sqlContext.read.format('csv').options(header='false', inferSchema='true').load('stopwords.txt')


st = [w[0] for w in stopwords.select("_c0").collect()]

#Function to tokenize words and remove stop words
def ret_token(string):
    if string is None:
        return None
    words = re.split(r'\W+', string)
    words = [word for word in words if word not in st]
    return words


```

Tokenzing Google and Amaozon datasets:

```python
google = google.filter(google.description.isNotNull())
google_tokens = google.rdd.map(lambda x:(x[0],ret_token(x[2].lower())))
google_tokens_df = spark.createDataFrame(google_tokens, ["id", "value"])

google_tokens_df.show(20)
```

```python
amazon = amazon.filter(amazon.description.isNotNull())
amazon_tokens = amazon.rdd.map(lambda x:(x[0],ret_token(x[2].lower())))
amazon_tokens_df = spark.createDataFrame(amazon_tokens, ["id", "value"])

amazon_tokens_df.show(20)
```

<h3> Part 3: </h3> Write a function that takes a list of tokens and returns a dictionary mapping tokens to term frequency.

```python
from pyspark.mllib.feature import HashingTF, IDF

def term_freq(tokens):
    dict_tokens = {}
    for t in tokens:
        if len(t) > 0:
            if t in dict_tokens:
                dict_tokens[t]=dict_tokens[t]+1
            else:
                dict_tokens[t]=1
    return dict_tokens
    
google_terms_freq=google_tokens.map(lambda x:(x[0],term_freq(x[1])))
#print(google_terms_freq.take(2))
google_terms = spark.createDataFrame(google_terms_freq, ["id", "value"])
google_terms.show(5)

amazon_terms_freq=amazon_tokens.map(lambda x:(x[0],term_freq(x[1])))
amazon_terms = spark.createDataFrame(amazon_terms_freq, ["id", "value"])
amazon_terms.show(5)
```

<h3> Part 4: </h3> Combine the datasets to create a corpus.

```python
corpus=google_tokens_df.union(amazon_tokens_df)

corpus.take(4)
```

<h3> Part 5: </h3> Write an IDF function

```python

n = corpus.count()
dict_total={}

def idf(col):
    for each in col['value']:
        for w in set(each):
            if len(w)>0:
                if w in dict_total:
                    dict_total[w]=dict_total[w]+1
                else:
                    dict_total[w]=1
    for i in dict_total:
        dict_total[i]=n/dict_total[i]
    return dict_total

col=corpus.select('value')
col=col.toPandas()
idf(col)
```

Plotting a histogram of IDF values

```python
import matplotlib.pyplot as plt
p=[dict_total[i] for i in dict_total]
fig, ax = plt.subplots()

num_bins = 50
n, bins, patches = plt.hist(p, num_bins, facecolor='maroon')
plt.title('Histogram of IDF values')
plt.xlabel('IDF values')
plt.ylabel('Frequency')
plt.show()

```
