<h3> Part 1: </h3> Data Loading
    
    

```python
from pyspark import SparkContext
from pyspark.ml.linalg import DenseVector
from pyspark.ml.feature import StandardScaler

sqlContext = SQLContext(spark)


```

```python
filename="default of credit card clients.csv"
spark = SparkSession.builder.getOrCreate()
data = spark.read.load(filename , format = 'csv',header = 'True', inferschema = 'true', sep= ",")
print("# of rows in data:",data.count())

# all your data are belong to Spark DataFrame
data.printSchema()
```

```python
Cols = []
from pyspark.sql.functions import col , column
for i in data.columns:
    Cols.append(i)
    data = data.withColumn(i, col(i).cast("double"))
Cols=Cols[:len(Cols)-2]
Cols=Cols[1:]
print(Cols)
data.toPandas().info()
```

```python
data=data.filter(data.X1.isNotNull())
data.show(3)
```

<h3> Part 2: </h3> Data Transformation

```python
data.describe().toPandas().transpose()
```

EDUCATION (X3) and MARITAL STATUS (X4) has few missing values as 0; Replaced them with mode of the column

```python

print(data.toPandas()['X3'].unique())
data.toPandas()['X3'].unique()
```

```python
from pyspark.sql.functions import when
data = data.withColumn("X3", when(data["X3"] == 0.0, 6.0).otherwise(data["X3"]))
data = data.withColumn("X4", when(data["X4"] == 0.0, 3.0).otherwise(data["X4"]))
print(data.toPandas()['X3'].unique())
data.toPandas()['X3'].unique()
```

<h3> Machine Learning Pipeline </h3> 
* Features to vector transformation
* Normalizing Features

```python
from pyspark.ml.feature import VectorAssembler
stages = []

assembler = VectorAssembler(inputCols=Cols, outputCol="features")
stages += [assembler]

standardScaler = StandardScaler(inputCol="features", outputCol="features_scaled")
stages += [standardScaler]

from pyspark.ml import Pipeline
pipeline = Pipeline(stages = stages)


pipelineModel = pipeline.fit(data)
print(pipelineModel)
df = pipelineModel.transform(data)

selectedCols = ['Y', 'features_scaled']
df = df.select(selectedCols)
df.printSchema()
```

```python
df.show(3)
```

Split dataset into training and test set

```python
df=df.withColumnRenamed('Y', 'label')
df=df.withColumnRenamed('features_scaled', 'features')
train, test = df.randomSplit([0.7, 0.3], seed = 1234)
print("Training Dataset Count: ", train.count())
print("Test Dataset Count: ", test.count())
```

<h3> Part 3 and 4: </h3> Model Learning and Evaluation using Cross Validation

```python
#References: https://docs.databricks.com/spark/latest/mllib/binary-classification-mllib-pipelines.html
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator
evaluator = BinaryClassificationEvaluator()
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.classification import GBTClassifier
```

<h3> Logistic Regression: </h3> Model Learning and Evaluation

```python

lr = LogisticRegression(labelCol="label", featuresCol="features")
paramGrid = (ParamGridBuilder()\
             .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0])\
             .addGrid(lr.maxIter, [1, 5, 10,100])\
             .build())

crossval = CrossValidator(estimator=lr,estimatorParamMaps=paramGrid,numFolds=5,evaluator=evaluator) 
cvModel = crossval.fit(train)
predictions = cvModel.transform(test)

print('The accuracy of the model on the test data is:', evaluator.evaluate(predictions))
```

<h3> Random Forest: </h3> Model Learning and Evaluation

```python
rf = RandomForestClassifier(featuresCol = 'features', labelCol = 'label')
paramGrid = (ParamGridBuilder()\
             .addGrid(rf.maxDepth, [2, 4, 6,8])
             .addGrid(rf.maxBins, [20, 60])
             .addGrid(rf.numTrees, [5, 20,10,40])
             .build())

crossval = CrossValidator(estimator=rf,estimatorParamMaps=paramGrid,numFolds=5,evaluator=evaluator) 
cvModel = crossval.fit(train)
predictions = cvModel.transform(test)
print('The accuracy of the model on the test data is:', evaluator.evaluate(predictions))
```

<h3> Gradient Boosting: </h3> Model Learning and Evaluation

```python
gbt = GBTClassifier(labelCol="label", featuresCol="features", maxIter=10)
paramGrid = (ParamGridBuilder()
             .addGrid(gbt.maxDepth, [2, 4, 6])
             .addGrid(gbt.maxBins, [20, 60])
             .addGrid(gbt.maxIter, [10, 20])
             .build())
crossval = CrossValidator(estimator=gbt,estimatorParamMaps=paramGrid,numFolds=5,evaluator=evaluator) 
cvModel = crossval.fit(train)
predictions = cvModel.transform(test)
print('The accuracy of the model on the test data is:', evaluator.evaluate(predictions))
```
